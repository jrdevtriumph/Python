```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import seaborn as sns

# Generating synthetic customer data (Age, Spend Amount, Visit Frequency)
np.random.seed(42)
customer_data = {
    'Age': np.random.randint(18, 60, 100),  # Age of customers
    'Annual_Spend': np.random.normal(500, 150, 100),  # Annual spend of customers
    'Visit_Frequency': np.random.randint(1, 15, 100)  # Frequency of visits per year
}

# Create a DataFrame from the synthetic customer data
df_customers = pd.DataFrame(customer_data)

# Standardize the data for better clustering (important for KMeans)
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_customers)

# Apply KMeans clustering (dividing into 3 clusters)
kmeans = KMeans(n_clusters=3, random_state=42)
df_customers['Segment'] = kmeans.fit_predict(df_scaled)

# Visualizing the clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Annual_Spend', y='Visit_Frequency', hue='Segment', data=df_customers, palette='Set1', s=100)
plt.title('Customer Segmentation')
plt.xlabel('Annual Spend ($)')
plt.ylabel('Visit Frequency')
plt.legend(title='Customer Segment')
plt.show()
```
