```python
# Import necessary libraries for linear regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error

# Generating synthetic sales data for training the model
np.random.seed(42)
data = {
    'Price': np.random.randint(10, 100, 30),  # Product price
    'Discount': np.random.randint(0, 30, 30),  # Discount percentage
    'Advertising_Spend': np.random.randint(500, 5000, 30),  # Advertising spend
    'Sales': np.random.randint(50, 500, 30)  # Units sold
}

# Create a DataFrame with the synthetic data
df_sales_predict = pd.DataFrame(data)

# Features (independent variables): Price, Discount, Advertising Spend
X = df_sales_predict[['Price', 'Discount', 'Advertising_Spend']]

# Target (dependent variable): Sales
y = df_sales_predict['Sales']

# Train-Test Split to evaluate the model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.8, random_state=42)

# Apply Linear Regression
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate model performance using Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)
print(f'Mean Absolute Error: {mae}')

# Plotting Actual vs Predicted Sales
plt.figure(figsize=(10, 6))
plt.scatter(y_test, y_pred)
plt.plot([0, 500], [0, 500], '--r', label="Perfect Prediction")
plt.title('Actual vs Predicted Sales')
plt.xlabel('Actual Sales')
plt.ylabel('Predicted Sales')
plt.legend()
plt.show()
```
