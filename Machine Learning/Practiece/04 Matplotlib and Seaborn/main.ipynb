```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample data
scores = {
    'Math': [90, 85, 78, 92, 88],
    'Science': [87, 89, 94, 86, 82],
    'English': [78, 85, 88, 80, 92],
    'Bangla': [20, 55, 77, 25, 14]
}
students = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']

df_scores = pd.DataFrame(scores, index=students)

# Plot a bar chart of average scores
df_scores['Average'] = df_scores.mean(axis=1)
df_scores['Average'].plot(kind='bar', color='skyblue', title='Average Scores by Student')
plt.ylabel('Average Score')
plt.show()
```

```python
# Heatmap of scores
plt.figure(figsize=(8, 6))
sns.heatmap(df_scores[scores.keys()], annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Heatmap of Scores")
plt.show()
```
