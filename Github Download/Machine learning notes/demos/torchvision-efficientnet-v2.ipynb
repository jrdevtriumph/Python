# Using EfficientNetV2 in PyTorch and TorchVision

TorchVision v 0.13 [added EfficientNetV2](https://pytorch.org/vision/main/models/generated/torchvision.models.efficientnet_v2_l.html#torchvision.models.efficientnet_v2_l) to its library along with ImageNet weights to use it as a pre-trained model.

Here is a quick usage example.

```python
%load_ext watermark
%watermark -p torch,torchvision
```

```python
from torchvision.models import efficientnet_v2_l, EfficientNet_V2_L_Weights

model = efficientnet_v2_l(weights=EfficientNet_V2_L_Weights.IMAGENET1K_V1)
model.eval();
```

```python
from PIL import Image
import matplotlib.pyplot as plt

img = Image.open('data/cat/cat.jpeg')
plt.imshow(img);
```

```python
import torch

preprocessor = EfficientNet_V2_L_Weights.IMAGENET1K_V1.transforms()
img_t = preprocessor(img)

logits = model(img_t.unsqueeze(0))
label = torch.argmax(logits).item()
label
```
